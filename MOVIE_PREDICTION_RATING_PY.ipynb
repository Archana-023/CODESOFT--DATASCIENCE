```python
dataset.dropna(inplace=True)
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset['Year'] = dataset['Year'].str.extract('(\d+)')
dataset['Year'] = pd.to_numeric(dataset['Year'], errors='coerce')
```

```python
dataset['Duration'] = dataset['Duration'].str.extract('(\d+)')
dataset['Duration'] = pd.to_numeric(dataset['Duration'], errors='coerce')
```

```python
dataset["Year"].head()
```

```python
genre=dataset['Genre']
genre.head(5)
```

```python
genres=dataset['Genre'].str.split(',',expand=True)
genres.head(5)
```

```python
genre_counts = {}
for genre in genres.values.flatten():
    if genre is not None:
        if genre in genre_counts:
            genre_counts[genre] += 1
        else:
            genre_counts[genre] = 1

genereCounts = {genre: count for genre, count in sorted(genre_counts.items())}
for genre, count in genereCounts.items():
    print(f"{genre}: {count}")
```

**VISUALIZATION OF DATA FROM THE GIVEN DATASET OF MOVIE PREDICTION RATING**

```python
import seaborn as sb
import plotly.express as px
import matplotlib.pyplot as mpl
from wordcloud import WordCloud
```

```python
sb.set(style = "ticks")
```

```python
ax = sb.boxplot(data=datset, y='Duration')
ax.set_title("Box Plot of Average Movie Durations")
ax.set_ylabel('Average Duration(in minutes)')
mpl.show()
```

```python
genre_counts = datset['Genre'].str.split(', ', expand=True).stack().value_counts()

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(genre_counts)

mpl.figure(figsize=(10, 6))
mpl.imshow(wordcloud, interpolation='bilinear')
mpl.axis('off')
mpl.title('Genre Word Cloud')
mpl.show()
```

```python
ax = sb.histplot(data = datset, x = "Rating", bins = 20, kde = True)
ax.set_xlabel('Rating')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Movie Ratings')
mpl.show()
```

```python
ax = sb.boxplot(data=datset, y='Rating')
ax.set_ylabel('Rating')
ax.set_title('Box Plot of Movie Ratings')
mpl.show()
```

```python
Q1 = datset['Rating'].quantile(0.25)
Q3 = datset['Rating'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
datset = datset[(datset['Rating'] >= lower_bound) & (datset['Rating'] <= upper_bound)]
datset.head(5)
```
